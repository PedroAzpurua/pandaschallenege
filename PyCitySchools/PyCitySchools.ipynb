```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# 1. Calculate the total number of scchools 
total_schools_df = len(school_data["School ID"].unique())
    #Initially did with count but this would prevent duplicate schools. 
total_schools_df
```

```python
#2. Total number of students
#total_students = student_data["Student ID"].count()
    #Same as above, went with this count method to prevent duplicates
total_students_df = len(student_data["Student ID"].unique())
total_students_df
```

```python
#3. Calculate total budget
total_budget_df = school_data["budget"].sum()
total_budget_df
```

```python
#4. Calculate the average Maths score
average_math_score_df = student_data["maths_score"].mean()
average_math_score_df
```

```python
#5 Calculate the average Reading Score
average_reading_score_df = student_data["reading_score"].mean()
average_reading_score_df
```

```python
#6. Calculate the number of students who passed maths
math_pass = student_data.loc[student_data["maths_score"]>=50]["maths_score"].count()
math_pass_percentage = (math_pass/total_students_df)*100
math_pass_percentage 
```

```python
#7. Calculate the number of students who passed reading
reading_pass = student_data.loc[student_data["reading_score"]>=50]["reading_score"].count()
reading_pass_percentage = (reading_pass/total_students_df)*100
reading_pass_percentage 
```

```python
#8 Calc number of students who passed maths and reading 
passed_both = student_data[(student_data["maths_score"] >=50) & (student_data["reading_score"] >=50)]["student_name"].count()
passed_both_percentage =(passed_both/total_students_df)*100
passed_both_percentage
```

```python
#9 Create a data frame to hold the above results
student_results_df = pd.DataFrame({
    "Total Schools" :[total_schools_df],
    "Total Students":[total_students_df],
    "Total Budget":[total_budget_df],
    "Average Math Results":[average_math_score_df],
    "Average Reading Result" :[average_reading_score_df],
    "Percentage of Students Passing Maths" :[math_pass_percentage],
    "Percentage of Students Passing Reading":[reading_pass_percentage],
    "Precentage of Students Passing Both Units" :[passed_both_percentage]

})
student_results_df
```

```python
#10 Create a clean data frame to hold the above results
student_results_df.style.format({
    "Total Schools" : "{:,}", 
    "Total Students":"{:,}", 
    "Total Budget": "${:,}",
    "Average Math Results":"{:.2f}%", 
    "Average Reading Result" :"{:.2f}%", 
    "Percentage of Students Passing Maths" :"{:.2f}%", 
    "Percentage of Students Passing Reading":"{:.2f}%",
    "Precentage of Students Passing Both Units" :"{:.2f}%"
})


```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name

* Create a dataframe to hold the above results

```python
# Reviewing the structure of school_data_complete merged dataset
school_data_complete
```

```python
#1. Create an overview table that sumamrises School name
school_name_df = school_data_complete.set_index("school_name").groupby(["school_name"])
school_name_df

```

```python
#2. Create an overview table that sumamrises School type
school_type = school_data.set_index("school_name")["type"]
school_type
```

```python
#3. Count Students per school
students_per_school_df = school_name_df["Student ID"].count()
students_per_school_df
```

```python
#4. School Budgets
school_budget = school_data.set_index("school_name")["budget"]
school_budget
```

```python
#4. School Budgets per students
per_student_budget = school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"]
per_student_budget
```

```python
#5. School Average Maths Scores
per_school_math_avg = school_name_df["maths_score"].mean()
per_school_math_avg
```

```python
#6. School average reading scores
per_school_reading_avg = school_name_df["reading_score"].mean()
per_school_reading_avg
```

```python
#7. School Maths Passing %
school_pass_maths = school_data_complete[school_data_complete["maths_score"] >= 50].groupby("school_name")["Student ID"].count()/students_per_school_df*100
school_pass_maths
```

```python
#8. School Reading Passing %
school_pass_reading = school_data_complete[school_data_complete["reading_score"] >= 50].groupby("school_name")["Student ID"].count()/students_per_school_df*100
school_pass_reading
```

```python
#9. Passing both
school_pass_both = school_data_complete[(school_data_complete["reading_score"] >= 50) & (school_data_complete["maths_score"] >= 50)].groupby("school_name")["Student ID"].count()/students_per_school_df*100

school_pass_both
```

```python
#Creating a summary data set per school
per_school_summary_df = pd.DataFrame({
    "School Type": school_type,
    "Number of Students": students_per_school_df,
    "Total School Budget": school_budget,
    "Budget per Student": per_student_budget,
    "Average Maths Score": per_school_math_avg,
    "Average Reading Score": per_school_reading_avg,
    "Percentage of students passing Maths": school_pass_maths,
    "Percentage of students passing Reading": school_pass_reading,
    "Percentage of Students passing both": school_pass_both
})
per_school_summary_df
```
